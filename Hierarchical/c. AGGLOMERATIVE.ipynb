```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
```

```python
df = pd.read_excel('Data Test.xlsx')
df.head(10)
```

```python
data = df[['Tọa độ X', 'Tọa độ Y']].values
data
```

```python
# Khởi tạo thuật toán Agglomerative Clustering
n_clusters = 5  # Số cụm
linkage_method = 'average'  # Tùy chỉnh phương thức liên kết: 'single', 'complete', 'average', 'ward'
agglo_model = AgglomerativeClustering(n_clusters=n_clusters, linkage=linkage_method)

# Chạy thuật toán Agglomerative Clustering với dữ liệu trên
labels = agglo_model.fit_predict(data)
```

```python
# Vẽ biểu đồ phân cụm
plt.figure(figsize=(10, 5))
colors = plt.cm.get_cmap('tab20', n_clusters)  # Số lượng màu bằng số cụm
scatter = plt.scatter(data[:, 0], data[:, 1], c=labels, cmap=colors, marker='o', s=100)

plt.title(f'Agglomerative ({linkage_method.capitalize()} Linkage)')
plt.xlabel('Tọa độ X')
plt.ylabel('Tọa độ Y')

cbar = plt.colorbar(scatter, ticks=np.arange(n_clusters), aspect=10)
cbar.set_label('Cluster Label')

cbar.set_ticks(np.arange(n_clusters))
cbar.set_ticklabels([f'{i + 1}' for i in range(n_clusters)])
plt.show()
```

```python
# Chạy với tập dữ liệu lớn:
np.random.seed(42)
cluster_1 = np.random.normal(loc=[4, 4], scale=2.0, size=(5000, 2))
cluster_2 = np.random.normal(loc=[10, 10], scale=2.0, size=(5000, 2))
data = np.vstack([cluster_1, cluster_2])

n_clusters = 5  # Số cụm
linkage_method = 'average'  # Tùy chỉnh phương thức liên kết: 'single', 'complete', 'average', 'ward'
agglo_model = AgglomerativeClustering(n_clusters=n_clusters, linkage=linkage_method)

# Chạy thuật toán Agglomerative Clustering với dữ liệu trên
labels = agglo_model.fit_predict(data)

# Vẽ biểu đồ phân cụm
plt.figure(figsize=(10, 5))
colors = plt.cm.get_cmap('tab20', n_clusters)  # Số lượng màu bằng số cụm
scatter = plt.scatter(data[:, 0], data[:, 1], c=labels, cmap=colors, marker='o', s=100)

plt.title(f'Agglomerative ({linkage_method.capitalize()} Linkage)')
plt.xlabel('Tọa độ X')
plt.ylabel('Tọa độ Y')

cbar = plt.colorbar(scatter, ticks=np.arange(n_clusters), aspect=10)
cbar.set_label('Cluster Label')

cbar.set_ticks(np.arange(n_clusters))
cbar.set_ticklabels([f'{i + 1}' for i in range(n_clusters)])
plt.show()
```
